```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display, Math, Latex
#from sympy import symbols, exp, integrate, oo, simplify
#import sympy as sp

plt.rcParams['text.usetex'] = False

#plt.style.use('seaborn-v0_8')
```

```python
# get data and make dataset
# dataset = np.array([x1,y1],[x2,y2],...,[xi,yi])
np.random.seed(None) # Seed generator , None => system clock
jmax = 50
x = 0.0
y = 0.0 # Start at origin

x_list = [x]
y_list = [y]

for i in range(0, jmax + 1):
    x += (np.random.random() - 0.5) * 2.0 # -1 =< x =< 1
    y += (np.random.random() - 0.5) * 2.0 # -1 =< y =< 1
    x_list.append(x)
    y_list.append(y)

plt.plot(x_list, y_list, color='blue', marker='o')
plt.title('Random Walk')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.grid(True)
plt.show()
```
