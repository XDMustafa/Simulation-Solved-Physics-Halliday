```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display, Math, Latex
from sympy import symbols, exp, integrate, oo, simplify
import sympy as sp

plt.rcParams['text.usetex'] = True

plt.style.use('seaborn-v0_8')
```

```python
# filed electron
epcicon = 8.85 * 10**-12
k = 1 / (4*np.pi * epcicon)
r = 2
q = 10
E = k * (q /r**2)

x = np.linspace(0, r, 20)
y = np.linspace(0, r, 20)
X, Y = np.meshgrid(x, y)

A_X = E
A_Y = np.zeros_like(Y)

plt.figure()
plt.xlabel('x')
plt.ylabel('y')
plt.quiver(X, Y, A_X, A_Y, color='black')
plt.axis('equal')
plt.show()
```

```python
R = np.linspace(0,10,100)
Eplot = k * (q / R**2)
plt.plot(R,Eplot)
```
